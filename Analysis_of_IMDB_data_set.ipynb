## create separate column name as gender mapping to fill the null values

```python
# gender columns has some null value rest all other columns have no null values
movies_data["gender"].unique()# there is nan value in gender column
```

```python
movies_data["gender"].value_counts()
```

```python
# crate a function to encode gender column
def encode(X):
  if X=="Female":
    return 1
  else:
    return 0
```

```python
# use apply function to apply your function for encoding in gender column
movies_data["gender_mapping"] = movies_data["gender"].apply(encode)
```

```python
movies_data["gender_mapping"].value_counts() # we encode null value to mode of the value in gender column
```

# Q 9 Most productive director

```python
# group the data using director name and then get all title of each director and count them
movies_data.groupby("director_name")["title"].count().sort_values(ascending=False)
```

```python
# get the min and max year or director and count of there movies
data_agg = movies_data.groupby("director_name")[["year","title"]].aggregate({"year":['min','max'],'title':'count'})
data_agg
```

```python
data_agg.columns
```

```python
# join columns name
data_agg.columns=["_".join(i) for i in data_agg.columns]
```

```python
data_agg.head()
```

```python
data_agg.reset_index().head()
```

```python
# find years in which director made the movies
data_agg['yrs_active'] = data_agg['year_max'] - data_agg['year_min']
data_agg.head()
```

```python
# find movies per year
data_agg["movie_per_yr"] = data_agg['title_count']/data_agg['yrs_active']
data_agg.head()
```

```python
# find the director name and avg no of movie made in one year
data_agg.sort_values("movie_per_yr", ascending=False).head()
```

# **Q 10 Find Movie which earned highest profit**

```python
# creat function to find profit per movie
def prof(x):
  return x["revenue"] - x["budget"]
```

```python
movies_data["profit"] = movies_data[["revenue","budget"]].apply(prof,axis=1)
movies_data.head(3)
```

```python
# top 3 movies which has highest profit
top3_movies=movies_data.sort_values("profit",ascending = False)["title"].head(3)
```

```python
top3_movies.unique()
```
